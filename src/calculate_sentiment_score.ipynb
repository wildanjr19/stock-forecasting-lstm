```python
import pandas as pd
```

```python
data = pd.read_csv(r"F:\Willy's Project\Project\Sentiment Analysis and Stock Forecasting\data\numeric_sentiment.csv")
data
```

```python
# jadikan tangal menjadi datetime dengan format dd/mm/yyyy
data['Tanggal'] = pd.to_datetime(data['Tanggal'], format='%d/%m/%Y')
```

```python
# cek kembali tipe data
data.info()
```

```python
# hitung rata-rata sentimen per-hari
df_sentiment_daily = data.groupby('Tanggal')['Sentimen'].mean().reset_index()
```

```python
# shft satu langkah ke depan, dengan membuat kolom baru
df_sentiment_daily['sentiment_score_lag1'] = df_sentiment_daily['Sentimen'].shift(1)
```

```python
df_sentiment_daily
```

```python
# karena kolom 'sentiment_score_lag1' memiliki nilai NaN pada baris pertama, kita ganti dengan 0
df_sentiment_daily['sentiment_score_lag1'].fillna(0, inplace=True)
```

```python
df_sentiment_daily
```

```python
# bulatkan nilai agar mudah dibaca
df_sentiment_daily['sentiment_score_lag1'] = df_sentiment_daily['sentiment_score_lag1'].round(2)
```

```python
df_sentiment_daily
```

```python
# ubah nama kolom sentiment_score_lag1 menjadi sentiment_score
df_sentiment_daily.rename(columns={'sentiment_score_lag1': 'Sentiment_Score'}, inplace=True)
```

```python
# ambil hanya kolom 'Tanggal' dan 'sentiment_score'
df_sentiment_daily = df_sentiment_daily[['Tanggal', 'Sentiment_Score']]
```

```python
# simpan file ke CSV
df_sentiment_daily.to_csv(r"F:\Willy's Project\Project\Sentiment Analysis and Stock Forecasting\data\daily_sentiment.csv", index=False)
```
